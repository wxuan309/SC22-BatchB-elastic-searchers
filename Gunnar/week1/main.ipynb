#### Import packages

```python
import pandas as pd
import plotly.express as px
```

#### Load Dataset

```python
gamesales = pd.read_csv('./vgsales[1].csv')
```

#### Data Cleaning

```python
gamesales.info
```

```python
gamesales.head()
```

```python
gamesales.tail()
```

```python
gamesales = gamesales.loc[gamesales['Global_Sales'] > 5]
```

```python
gamesales.head()
```

```python
gamesalesNA = gamesales[['Name', 'Platform', 'Year', 'Genre', 'Publisher', 'NA_Sales']]
gamesalesNA = gamesalesNA.sort_values(by = 'NA_Sales')
gamesalesNA = gamesalesNA.iloc[::-1]
gamesalesNA.head(2)
```

```python
moderngamesalesNA = gamesalesNA.loc[gamesalesNA['Year'] > 2010]
moderngamesalesNA.head(3)
```

```python
gamesalesEU = gamesales[['Name', 'Platform', 'Year', 'Genre', 'Publisher', 'EU_Sales']]
gamesalesEU = gamesalesEU.sort_values(by = 'EU_Sales')
gamesalesEU = gamesalesEU.iloc[::-1]
gamesalesEU.head(2)
```

```python
moderngamesalesEU = gamesalesEU.loc[gamesalesEU['Year'] > 2010]
moderngamesalesEU.head(3)
```

```python
gamesalesJP = gamesales[['Name', 'Platform', 'Year', 'Genre', 'Publisher', 'JP_Sales']]
gamesalesJP = gamesalesJP.sort_values(by = 'JP_Sales')
gamesalesJP = gamesalesJP.iloc[::-1]
gamesalesJP.head(2)
```

```python
moderngamesalesJP = gamesalesJP.loc[gamesalesJP['Year'] > 2010]
moderngamesalesJP.head(3)
```
